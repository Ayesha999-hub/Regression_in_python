```python
import pandas as pd
import matplotlib.pyplot as plt
#a
earth_temp_data = pd.read_csv('/earth_surface_temperatures.csv')
print("the data for earth surface temperature is...\n", earth_temp_data )
missing_values = earth_temp_data.isnull().sum()
print("missing values in the data are..'\n..",missing_values)
 # Remove rows with missing values
earth_temp_data.dropna(inplace=True)
# Fill missing values with the meanmean_value = df['ColumnName'].mean()
earth_temp_data['Month'].interpolate(method='linear', inplace=True)  # Interpolate missing values in a column
 # Verify that there are no missing values left
missing_values_after_handling = earth_temp_data.isnull().sum()
print("missing values are..\n",missing_values_after_handling)

#b
# Check data types of all columns
data_types = earth_temp_data.dtypes
earth_temp_data['Years']= earth_temp_data['Years'].astype(str)
# Check data types after conversion
data_types_after_conversion = earth_temp_data.dtypes
print("data types after conversion is...\n",data_types_after_conversion)

#c
earth_temp_data['Date'] = earth_temp_data['Month'].astype(str) + '-' + earth_temp_data['Years']
print("the new dataframe with new column is....\n",earth_temp_data )

#d
# Calculate the first and third quartiles
Q1 =earth_temp_data ['Temperature'].quantile(0.25)
Q3 = earth_temp_data['Temperature'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define thresholds for considering data points as outliers
lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR

# Find and investigate outliers
outliers = earth_temp_data[(earth_temp_data['Temperature'] < lower_threshold) | (earth_temp_data['Temperature'] > upper_threshold)]
print("the outlier is ..\n",outliers)

#e
mean = earth_temp_data['Temperature'].mean()
median= earth_temp_data['Temperature'].median()
std = earth_temp_data['Temperature'].std()
temperature = earth_temp_data['Temperature'].max() - earth_temp_data['Temperature'].min()
print("mean = ",mean)
print("median = ",median)
print("standard deviation = ",std)
print("temperture = ",temperature)



earth_temp_data['Date'] = pd.to_datetime(earth_temp_data['Date'])
earth_temp_data['Month'] = earth_temp_data['Date'].dt.month
mean_monthly_variation =earth_temp_data.groupby('Month')['Temperature'].mean()
median_monthly_variation = earth_temp_data.groupby('Month')['Temperature'].median()
std_deviation_monthly_variation = earth_temp_data.groupby('Month')['Temperature'].std()
monthly_variation_range =earth_temp_data.groupby('Month')['Temperature'].max() - earth_temp_data.groupby('Month')['Temperature'].min()
print("mean = ",mean_monthly_variation)
print("median = ",median_monthly_variation)
print("standard deviation = ",std_deviation_monthly_variation)
print("temperture = ",monthly_variation_range)


#f

# Group the data by 'Country' and calculate the average temperature
average_temperature_by_country = earth_temp_data.groupby('Country')['Temperature'].mean()

# Reset the index to make 'Country' a regular column in the resulting DataFrame
average_temperature_by_country = average_temperature_by_country.reset_index()

# Rename the columns for clarity
average_temperature_by_country.columns = ['Country', 'Average Temperature']

# Print the resulting DataFrame with countries and their average temperatures
print(average_temperature_by_country)

#g

df = earth_temp_data[['Years', 'Temperature']]
global_avg_temp = df.groupby('Years')['Temperature'].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(global_avg_temp['Years'], global_avg_temp['Temperature'], marker='o', linestyle='-', color='b')
plt.title('Global Temperature Trend (1848-2020)')
plt.xlabel('Years')
plt.ylabel('Average Temperature (°C)')
plt.grid(True)
plt.show()

#h

# Convert the 'Date' column to a datetime object
earth_temp_data['Date'] = pd.to_datetime(earth_temp_data['Date'])

# Extract month and year from the 'Date' column
earth_temp_data['Month'] = earth_temp_data['Date'].dt.month
earth_temp_data['Years'] = earth_temp_data['Date'].dt.year
monthly_temperatures = earth_temp_data.groupby(['Country', 'Month']).agg(
    Highest_Temperature=('Temperature', 'max'),
    Lowest_Temperature=('Temperature', 'min')).reset_index()
highest_temperatures = monthly_temperatures.groupby('Country')['Highest_Temperature'].idxmax()
lowest_temperatures = monthly_temperatures.groupby('Country')['Lowest_Temperature'].idxmin()
highest_temperatures_data = monthly_temperatures.loc[highest_temperatures]
lowest_temperatures_data = monthly_temperatures.loc[lowest_temperatures]
country_name = 'Your_Country_Name'
country_data = monthly_temperatures[monthly_temperatures['Country'] == country_name]
plt.plot(country_data['Month'], country_data['Highest_Temperature'])
plt.title(f'Highest Temperatures for {country_name}')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.show()

#j
countries = [' Afghanistan', ' Zimbabwe', 'Antarctica ', ' Tajikistan', 'Ukraine']
filtered_df = earth_temp_data[earth_temp_data['Country'].isin(countries)]
filtered_df['Years'] = pd.to_datetime(filtered_df['Years'])
# Group and aggregate data
grouped_df = filtered_df.groupby(['Country', filtered_df['Years'].dt.year])['Temperature'].mean().reset_index()
for country in countries:
    country_data = grouped_df[grouped_df['Country'] == country]
    plt.plot(country_data['Year'], country_data['Temperature'], label=country)

plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trends for Selected Countries')
plt.legend()
plt.show()

#k
correlation_coefficient = earth_temp_data['Temperature'].corr(earth_temp_data['Anomaly'])


plt.scatter.earth_temp_data(['Temperature'],earth_temp_data['Anomaly'])
plt.xlabel('Temperature')
plt.ylabel('Anomaly')
plt.title('Temperature vs. Anomaly Scatterplot')
plt.show()

#i
earth_temp_data['Date'] = pd.to_datetime(earth_temp_data['Years'].astype(str) + '-' + earth_temp_data['Month'].astype(str))

monthly_mean = earth_temp_data.groupby(earth_temp_data['Date'].dt.month)['Temperature Anomaly'].mean()
# Adjust this value based on your research
high_threshold = 1.0
low_threshold = -1.0

# Identify months with consistently high or low anomalies
high_anomaly_months = monthly_mean[monthly_mean > high_threshold].index
low_anomaly_months = monthly_mean[monthly_mean < low_threshold].index
print(high_anomaly_months)
print(low_anomaly_months)

```
